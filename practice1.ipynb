```python
# p.197

def bs(array,target,start,end):
  if end<start:
    return 'NO'

  pivot = (start+end)//2
  if array[pivot]==target:
    return 'YES'

  elif array[pivot]<target:
    return bs(array,target,pivot+1,end)

  elif array[pivot]>target:
    return bs(array,target,start,pivot-1)

def solution(n,array,targets):
  results = []
  array.sort()
  for target in targets:
    result = bs(array,target,0,n-1)
    results.append(result)
  return results

n = 5
array = [8,3,7,9,2]
target = [5,7,9]

print(solution(n,array,target))
```

```python
# p.259

def dijkstra(start,end,n,graph):
  distance = [int(1e9)] * n
  h = []
  heapq.heappush(h,(0,start))
  distance[start] = 0

  while h:
    curr_dist,curr_node = heapq.heappop(h)
    if curr_node==end:
      return curr_dist
    if curr_dist>distance[curr_node]:
      continue
    for next_node in graph[curr_node]:
      if distance[next_node]>curr_dist+1:
        distance[next_node] = distance[curr_node] + 1
        heapq.heappush(h,(distance[next_node],next_node))

  return -1

def solution(n,k,x,graph):
  d0 = dijkstra(0,k,n,graph)
  d1 = dijkstra(k,x,n,graph)

  if d0!=-1 and d1!=-1:
    result = d0 + d1
  else:
    result = -1

  return result

n,m = map(int,input().split())
graph = [[] for _ in range(n)]
for _ in range(m):
  a,b = map(int,input().split())
  graph[a-1].append(b-1)
  graph[b-1].append(a-1)
x,k = map(int,input().split())
x -= 1
k -= 1

print(solution(n,k,x,graph))
```

```python
# p.303

def solution(indegree,graph_back,graph_for,data):
  result = [int(1e9)] * n
  queue = deque()
  for i in range(n):
    if indegree[i]==0:
      queue.append(i)

  while queue:
    # 노드 꺼내서 결과 확정
    curr_node = queue.popleft()
    temp = [result[back_node] for back_node in graph_back[curr_node]]
    if temp:
      result[curr_node] = max(temp) + data[curr_node]
    else:
      result[curr_node] = data[curr_node]

    # 다음 노드 진입차수 감소 및 0 될 시 큐에 투입
    for next_node in graph_for[curr_node]:
      indegree[next_node] -= 1
      if indegree[next_node]==0:
        queue.append(next_node)

  return result

n = int(input())
indegree = [0] * n
graph_back = [[] for _ in range(n)]
graph_for = [[] for _ in range(n)]
data = [0] * n
for i in range(n):
  temp = list(map(int,input().split()))
  data[i] = temp[0]
  for k in temp[1:-1]:
    k -= 1
    graph_back[i].append(k)
    graph_for[k].append(i)
    indegree[i] += 1

print(solution(indegree,graph_back,graph_for,data))
```

```python
# p.315

def count_num(array,i):
  right = bisect_right(array,i)
  left = bisect_left(array,i)
  return right-left

def choose_two(n):
  return int(n*(n-1)/2)

def solution(n,m,array):
  array.sort()
  array = [i-1 for i in array]
  count = [-1] * m
  for i in range(m):
    count[i] = count_num(array,i)

  result = choose_two(n)

  for i in count:
    if i>1:
      result -= choose_two(i)

  return result

n,m = map(int,input().split())
array = list(map(int,input().split()))

print(solution(n,m,array))
```

```python
# p.316

def find_next(food_times,k,r):
  next = k%r
  count = -1
  for i,time in enumerate(food_times):
    if time>0:
      count += 1
      if count==next:
        return i+1

def solution(food_times,k):
  while True:
    temp = [i for i in food_times if i>0]
    if not temp:
      return -1
    m = min(temp)
    r = len(temp)
    if m*r<k:
      food_times = [i-m if i>0 else i for i in food_times]
      k -= m*r
    else:
      return find_next(food_times,k,r)

food_times = [8,6,4,3,2,2,3,5,3,1,4,4,6,2,5,2,2]
k = 20
print(solution(food_times,k))
```

```python
# p.323

def split_i(string,n,i):
  q = n//i
  r = n%i
  result = []
  for j in range(q):
    result.append(string[j*i:(j+1)*i])
  result.append(string[n-r:])

  return result

def merge(array):
  result = ''
  curr = ''
  count = 1
  for a in array:
    if a==curr:
      count += 1
    else:
      result += str(count)
      result += a
      curr = a
      count = 1

  temp = [a for a in result if a!='1']

  return len(temp)

def solution(string):
  n = len(string)
  result = int(1e9)
  for i in range(1,(n+1)//2+1):
    # i는 압축 단위길이
    # 문자열을 i개씩 묶어서 쪼개는 함수
    temp = split_i(string,n,i)
    # 쪼갠 문자열을 합친 길이를 반환하는 함수
    result = min(merge(temp),result)

  return result

string = 'xababcdcdababcdcd'
print(solution(string))
```

```python
# p.325

def f(key,n):
  result = copy.deepcopy(key)
  for i in range(n):
    for j in range(n):
      result[i][j] = key[n-1-j][i]

  return result

def rotate(key,i,n):
  result = copy.deepcopy(key)
  for _ in range(i):
    result = f(result,n)

  return result

def solution(key,lock):
  n = len(lock)
  m = len(key)
  k = n+2*m-2
  # n+2m-2 크기의 board 생성
  board = [[0]*k for _ in range(k)]
  # 보드 중간에 n*n 크기의 lock 입력
  for i in range(n):
    for j in range(n):
      board[i+m-1][j+m-1] = lock[i][j]
  # 모든 회전에 대하여
  for r in range(4):
    rotated_key = rotate(key,r,n)
    # 모든 칸에 대하여
    for i in range(k-m+1):
      for j in range(k-m+1):
        new_board = copy.deepcopy(board)
        # new_board에서 key 부분 합치기
        for a in range(m):
          for b in range(m):
            new_board[i+a][j+b] += rotated_key[a][b]
        # new_board의 lock 부분 다 1인지 확인
        check = True
        for x in range(m-1,n+m-1):
          if not check:
            break
          for y in range(m-1,n+m-1):
            if new_board[x][y]!=1:
              check = False
              break

        if check:
          return 'true'

  return 'false'

key = [[0,0,0],[1,0,0],[0,1,1]]
lock = [[1,1,1],[1,1,0],[1,0,1]]
print(solution(key,lock))
```

```python
# p.346

def rev(p):
  result = ''
  for a in p:
    if a==')':
      result += '('
    else:
      result += ')'
  return result

def check_alr(p):
  cnt_0 = 0
  cnt_1 = 0
  for a in p:
    if a=='(':
      cnt_0 += 1
    else:
      cnt_1 += 1
    if cnt_1>cnt_0:
      return False
  return True

def split_bal(p):
  u = ''
  cnt_0 = 0
  cnt_1 = 0
  for a in p:
    u += a
    if a=='(':
      cnt_0 += 1
    else:
      cnt_1 += 1
    if cnt_0==cnt_1:
      break

  v = p[cnt_0+cnt_1:]

  return u,v

def dfs(w):
  if not w:
    return ''

  u,v = split_bal(w)
  if check_alr(u):
    return u + dfs(v)
  else:
    return '(' + dfs(v) + ')' + rev(u[1:-1])

def solution(p):
  if check_alr(p):
    return p

  return dfs(p)

print(solution(input()))
```

```python
# p.361

def solution(n,stages):
  stages = [a-1 for a in stages]
  current_stage = [-1] * n
  stages.sort()
  for i in range(n):
    current_stage[i] = bisect_right(stages,i) - bisect_left(stages,i)

  failure = [-1] * n
  total = len(stages)
  for i in range(n):
    if total==0:
      failure[i] = 0
    else:
      failure[i] = (current_stage[i] / total, i)
      total -= current_stage[i]

  failure.sort(key=lambda x: (-x[0],x[1]))

  return [a[1]+1 for a in failure]

n = 4
stages = [4,4,4,4,4]
print(solution(n,stages))
```

```python
# p.402
dx = [-1,1,0,0]
dy = [0,0,-1,1]

def find_next(graph,shark_size,shark_loc):
  x,y = shark_loc
  dist = [[int(1e9)]*n for _ in range(n)]
  dist[x][y] = 0
  q = deque()
  q.append((x,y))
  fishes = []
  while q:
    x,y = q.popleft()
    for i in range(4):
      nx = x + dx[i]
      ny = y + dy[i]
      if nx>=0 and nx<n and ny>=0 and ny<n:
        if graph[nx][ny]==0 or graph[nx][ny]==shark_size:
          if dist[x][y]+1<dist[nx][ny]:
            dist[nx][ny] = dist[x][y] + 1
            q.append((nx,ny))
        elif graph[nx][ny]<shark_size:
          if dist[x][y]+1<dist[nx][ny]:
            dist[nx][ny] = dist[x][y] + 1
            q.append((nx,ny))
            fishes.append((dist[nx][ny],nx,ny))

  if fishes:
    fishes.sort()
    t,nx,ny = fishes[0]
  else:
    nx,ny,t = -1,-1,-1

  return nx,ny,t

def solution(n,graph):
  for i in range(n):
    for j in range(n):
      if graph[i][j]==9:
        shark_loc = (i,j)
        graph[i][j] = 0
  shark_size = 2
  count = 2
  time = 0
  while True:
    # graph 복사
    temp = copy.deepcopy(graph)
    # 다음 물고기 위치확인
    nx,ny,t = find_next(temp,shark_size,shark_loc)
    if nx==-1:
      break
    # 그래프에서 물고기 없애기
    graph[nx][ny] = 0
    # 이동한만큼 시간증가
    time += t
    # 크기증가
    count -= 1
    if count==0:
      shark_size += 1
      count = shark_size
    # 상어이동
    shark_loc = (nx,ny)

  return time

n = int(input())
graph = []
for _ in range(n):
  graph.append(list(map(int,input().split())))

print(solution(n,graph))
```

```python
# p.404

dx = [0,-1,-1,0,1,1,1,0,-1]
dy = [0,0,-1,-1,-1,0,1,1,1]

def find_i(graph,k):
  for i in range(4):
    for j in range(4):
      if graph[i][j][0]==k:
        return i,j,graph[i][j][1]

  return -1,-1,-1

def fish_move(graph,shark_loc):
  temp = copy.deepcopy(graph)
  for i in range(1,17):
    x,y,dir = find_i(temp,i)
    if x==-1:
      continue
    t = [j for j in range(dir,9)] + [j for j in range(1,dir)]
    for d in t:
      temp[x][y] = (temp[x][y][0],d)
      nx = x + dx[d]
      ny = y + dy[d]
      if nx>=0 and nx<4 and ny>=0 and ny<4:
        if (nx,ny)!=shark_loc:
          temp[x][y],temp[nx][ny] = temp[nx][ny],temp[x][y]
          break

  return temp

def shark_move(graph,shark_loc,shark_dir):
  next_sharks = []
  x,y = shark_loc
  while True:
    nx = x + dx[shark_dir]
    ny = y + dy[shark_dir]
    if nx>=0 and nx<4 and ny>=0 and ny<4:
      if graph[nx][ny]!=(0,0):
        next_sharks.append((nx,ny))
      x,y = nx,ny
    else: break

  return next_sharks


def dfs(graph,shark_loc,shark_dir,eat):
  global result
  temp = fish_move(graph,shark_loc)
  next_sharks = shark_move(temp,shark_loc,shark_dir)

  if next_sharks:
    for nx,ny in next_sharks:
      t = temp[nx][ny][0]
      next_shark_loc = (nx,ny)
      next_shark_dir = temp[nx][ny][1]
      temp_ = copy.deepcopy(temp)
      temp_[nx][ny] = (0,0)
      dfs(temp_,next_shark_loc,next_shark_dir,eat+t)

  if not next_sharks:
    result = max(result,eat)

def solution(graph):
  global result
  result = 0
  temp = copy.deepcopy(graph)
  shark_loc = (0,0)
  shark_dir = temp[0][0][1]
  temp[0][0] = (0,0)

  dfs(temp,shark_loc,shark_dir,graph[0][0][0])

  return result

graph = []
for _ in range(4):
  a0,a1,b0,b1,c0,c1,d0,d1 = map(int,input().split())
  graph.append([(a0,a1),(b0,b1),(c0,c1),(d0,d1)])

print(solution(graph))
```
