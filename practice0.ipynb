```python
# p.168

array = list(map(int,input().split()))

n = len(array)

def quick_sort(array,start,end):
  if start>=end:
    return

  pivot = start
  left = start+1
  right = end

  while left<right:
    while left<=end and array[left]<array[pivot]:
      left += 1
    while right>=start and array[right]>array[pivot]:
      right -= 1
    if left>right:
      array[pivot],array[right] = array[right],array[pivot]
    else:
      array[left],array[right] = array[right],array[left]

  quick_sort(array,start,right-1)
  quick_sort(array,right+1,end)

quick_sort(array,0,n-1)

print(array)
```

```python
# p.237

inf = int(1e9)

n,m = map(int,input().split())
start = int(input())

graph = [[] for _ in range(n+1)]

for _ in range(m):
  a,b,c = map(int,input().split())
  graph[a].append((b,c))

visited = [False] * (n+1)
table = [inf] * (n+1)

table[1] = 0

def set_start(visited,table):
  rep_table =[]
  for i in range(n+1):
    if not visited[i]:
      rep_table.append(table[i])
    else:
      rep_table.append(inf)

  return rep_table.index(min(rep_table))

for i in range(n):

  # 시작지점 설정
  start = set_start(visited,table)

  # 시작지점 방문처리
  visited[start] = True

  # 시작지점으로부터 거리갱신
  for dest,cost in graph[start]:
    dist = table[start] + cost
    if dist<table[dest]:
      table[dest] = dist

print(table)
```

```python
# p.335

def move(n,start,worker,weak,weak_label):
  # start로부터 worker만큼 움직이고 지나간 weak들은 weak_label true 처리하기

  new_weak_label = weak_label.copy()
  end = (start+worker)%n

  if end<start:
    for i,a in enumerate(weak):
      if a>=start or a<=end:
        new_weak_label[i] = True
  else:
    for i,a in enumerate(weak):
      if a>=start and a<=end:
        new_weak_label[i] = True

  return end,new_weak_label

def next_start(end,weak):
  if end<weak[0] or end>=weak[-1]:
    start = weak[0]
  else:
    for i in range(len(weak)-1):
      if end>=weak[i] and end<weak[i+1]:
        start = weak[i+1]
        break

  return start

def solution(n,weak,dist):
  m = len(weak)
  success = False

  for i in range(1,m+1):
    if success:
      break

    for start in weak:
      if success:
        break

      for workers in itertools.permutations(dist,i):
        weak_label = [False] * m
        for worker in workers:
          end,weak_label = move(n,start,worker,weak,weak_label)
          start = next_start(end,weak)
        if weak_label == [True] * m:
          result = i
          success = True
          break

  if not success:
    return -1

  return result

print(solution(12,[1,5,6,8,10],[1,2,3,4]))
print(solution(12,[1,3,4,9,10],[1,2]))
```

```python
# p.339

n,m,k,start = map(int,input().split())

graph = [[] for _ in range(n+1)]

for _ in range(m):
  a,b = map(int,input().split())
  graph[a].append(b)

dist = [int(1e9)] * (n+1)
dist[start] = 0

hq = []

heapq.heappush(hq,(0,start))

while hq:
  # 큐에서 최단 거리 노드 추출
  curr_dist, curr_node = heapq.heappop(hq)

  # 큐에서 꺼낸 노드의 거리정보가 거리 DB보다 크다면 무시
  if curr_dist > dist[curr_node]:
    continue

  # 해당 노드에서 갈 수 있는 노드 확인 후 갱신되는지 확인. 갱신된다면 거리 갱신하고 큐에 넣기
  for next_node in graph[curr_node]:
    if dist[next_node] > curr_dist + 1:
      dist[next_node] = curr_dist + 1
      heapq.heappush(hq,(curr_dist+1,next_node))

result = []
for i,d in enumerate(dist):
  if d==k:
    result.append(i)

if len(result)==0:
  print(-1)
else:
  for i in result:
    print(i)
```

```python
# p.346

p = input()

def check(p):
  count_0 = 0
  count_1 = 0

  result = True

  for a in p:
    if a==')':
      count_1 += 1
    else:
      count_0 += 1
    if count_0 < count_1:
      result = False

  return result

def uv_divide(p):
  u = ''
  count_0 = 0
  count_1 = 0

  for i,a in enumerate(p):
    u += a
    if a==')':
      count_1 += 1
    else:
      count_0 += 1
    if count_0==count_1:
      break

  v = p[i+1:]

  return u,v

def rv(u):
  reversed_u = ''
  for a in u:
    if a==')':
      reversed_u += '('
    else:
      reversed_u += ')'

  return reversed_u

def modify(p):
  # step 0
  if check(p):
    return p

  # step 1
  if len(p)==0:
    return p

  # step 2
  u,v = uv_divide(p)

  # step 3
  if check(u):
    v = modify(v)
    return u+v

  modified = ""
  # step 4
  if not check(u):
    # step 4-1
    modified += '('
    # step 4-2
    v = modify(v)
    modified += v
    # step 4-3
    modified += ')'
    # step 4-4
    modified += rv(u[1:-1])
    # step 4-5
    return modified

modify(p)
```

```python
# p.353

n,m,M = map(int,input().split())

graph = []

for _ in range(n):
  graph.append(list(map(int,input().split())))

def bfs(visited,start):
  # start부터 방문 안했고 조건 만족하는 node 방문처리
  queue = deque()
  queue.append(start)
  new_visited = visited.copy()
  group = []

  while queue:
    curr_node = queue.popleft()
    new_visited[curr_node[0]][curr_node[1]] = True
    group.append(curr_node)

    i,j = curr_node
    if i-1 in list(range(n)) and j in list(range(n)):
      if not new_visited[i-1][j] and m<=abs(graph[i-1][j]-graph[i][j])<=M:
        queue.append((i-1,j))
    if i+1 in list(range(n)) and j in list(range(n)):
      if not new_visited[i+1][j] and m<=abs(graph[i+1][j]-graph[i][j])<=M:
        queue.append((i+1,j))
    if i in list(range(n)) and j-1 in list(range(n)):
      if not new_visited[i][j-1] and m<=abs(graph[i][j-1]-graph[i][j])<=M:
        queue.append((i,j-1))
    if i in list(range(n)) and j+1 in list(range(n)):
      if not new_visited[i][j+1] and m<=abs(graph[i][j+1]-graph[i][j])<=M:
        queue.append((i,j+1))

  return new_visited,group

def divide(graph):
  groups = []
  visited = [[False]*n for _ in range(n)]

  for i in range(n):
    for j in range(n):
      if not visited[i][j]:
        visited,group = bfs(visited,(i,j))
        groups.append(group)

  return groups

def distribute(groups,graph):
  new_graph = copy.deepcopy(graph)
  for group in groups:
    count = 0
    summ = 0
    for i,j in group:
      count += 1
      summ += graph[i][j]
    for i,j in group:
      new_graph[i][j] = summ//count

  return new_graph

count = 0
while True:
  group = divide(graph)
  if len(group)==n**2:
    break
  graph = distribute(group,graph)
  count += 1

print(count)
```

```python
# p.353

def find_next(loc,visited,board):
  (i0,j0),(i1,j1) = list(loc)
  dx = [-1,1,0,0]
  dy = [0,0,-1,1]

  next_nodes = []

  for k in range(4):
    n_i0 = i0 + dx[k]
    n_j0 = j0 + dy[k]
    n_i1 = i1 + dx[k]
    n_j1 = j1 + dy[k]

    next_loc = {(n_i0,n_j0),(n_i1,n_j1)}

    if next_loc not in visited and board[n_i0][n_j0]==0 and board[n_i1][n_j1]==0:
      next_nodes.append(next_loc)

  # 누워있을 때
  if i0==i1:
    if board[i0-1][j0]==0 and board[i1-1][j1]==0:
      if {(i0-1,j0),(i0,j0)} not in visited:
        next_nodes.append({(i0-1,j0),(i0,j0)})
      if {(i1-1,j1),(i1,j1)} not in visited:
        next_nodes.append({(i1-1,j1),(i1,j1)})
    if board[i0+1][j0]==0 and board[i1+1][j1]==0:
      if {(i0+1,j0),(i0,j0)} not in visited:
        next_nodes.append({(i0+1,j0),(i0,j0)})
      if {(i1+1,j1),(i1,j1)} not in visited:
        next_nodes.append({(i1+1,j1),(i1,j1)})

  # 서있을 때
  else:
    if board[i0][j0-1]==0 and board[i1][j1-1]==0:
      if {(i0,j0),(i0,j0-1)} not in visited:
        next_nodes.append({(i0,j0),(i0,j0-1)})
      if {(i1,j1),(i1,j1-1)} not in visited:
        next_nodes.append({(i1,j1),(i1,j1-1)})
    if board[i0][j0+1]==0 and board[i1][j1+1]==0:
      if {(i0,j0),(i0,j0+1)} not in visited:
        next_nodes.append({(i0,j0),(i0,j0+1)})
      if {(i1,j1),(i1,j1+1)} not in visited:
        next_nodes.append({(i1,j1),(i1,j1+1)})

  return next_nodes

def solution(board):
  n = len(board)
  queue = deque()
  start = {(1,1),(1,2)}
  queue.append((start,0))
  visited = []
  new_board = [[1]*(n+2) for _ in range(n+2)]
  for i in range(n):
    for j in range(n):
      new_board[i+1][j+1] = board[i][j]

  while queue:
    curr_loc,curr_dist = queue.popleft()
    if curr_loc in visited:
      continue

    if (n,n) in curr_loc:
      return curr_dist

    visited.append(curr_loc)
    next_nodes = find_next(curr_loc,visited,new_board)
    for next_loc in next_nodes:
      queue.append((next_loc,curr_dist+1))

board = [[0,0,0,1,1],[0,0,0,1,0],[0,1,0,1,1],[1,1,0,0,1],[0,0,0,0,0]]
print(solution(board))
```

```python
# p.361

def f(stages,i,start):
  j = start
  while j<len(stages) and stages[j]==i:
    j += 1

  if len(stages)==start:
    failure = 0
  else:
    failure = (j-start)/(len(stages)-start)
  next = j

  return failure, next

def solution(N,stages):
  failure_list = []
  stages.sort()

  start = 0
  for i in range(1,N+1):
    failure,start = f(stages,i,start)
    failure_list.append((failure,i))

  failure_list.sort(key=lambda x:(-x[0],x[1]))
  result = [data[1] for data in failure_list]

  return result

print(solution(5,[2,1,2,6,2,4,3,3]))
print(solution(4,[4,4,4,4,4]))
```

```python
# p.369

n,c = map(int,input().split())

data = []

for _ in range(n):
  data.append(int(input()))

data.sort()

def check(pivot):
  i = 1
  prev = data[0]
  result = True

  while i<pivot:
    curr = prev + pivot
    if curr in data:
      prev = curr
      i += 1
    else:
      idx = bisect_left(data,curr)
      if idx==n:
        result = False
        break
      else:
        prev = data[idx]
        i += 1

  return result

def bs(start,end):
  pivot = ((start+end+1)//2)

  if end==start:
    return start

  elif end-start==1:
    if check(pivot):
      return end
    else:
      return start

  if check(pivot):
    return bs(pivot,end)
  else:
    return bs(start,pivot-1)

print(bs(1,data[-1]-data[0]))
```

```python
# p.375

t = int(input())

def dp(prev_data,next_col):
  n = len(prev_data)
  result = []

  for i in range(n):
    if i==0:
      temp = max(prev_data[0]+next_col[0],prev_data[1]+next_col[0])
      result.append(temp)
    elif i==n-1:
      temp = max(prev_data[n-2]+next_col[n-1],prev_data[n-1]+next_col[n-1])
      result.append(temp)
    else:
      temp = max(prev_data[i-1]+next_col[i],prev_data[i]+next_col[i],prev_data[i+1]+next_col[i])
      result.append(temp)

  return result

for _ in range(t):
  n,m = map(int,input().split())
  graph = [[0]*m for _ in range(n)]
  a = list(map(int,input().split()))
  for i in range(len(a)):
    q = i//m
    r = i%m
    graph[q][r] = a[i]
  data = [[0]*m for _ in range(n)]
  for i in range(n):
    data[i][0] = graph[i][0]

  for j in range(m-1):
    prev_data = [row[j] for row in data]
    next_col = [row[j+1] for row in graph]
    col_data = dp(prev_data,next_col)
    for i in range(n):
      data[i][j+1] = col_data[i]

  last_col = [row[-1] for row in data]
  print(max(last_col))
```

```python
# p.407

def shark_move():
  global shark_info
  next_shark_info = []

  for shark_num,(shark_loc,curr_dir) in enumerate(shark_info):
    if shark_loc==-1:
      next_shark_info.append((-1,-1))
    else:
      x,y = shark_loc
      move = False
      for next_dir in dir_priority[shark_num][curr_dir]:
        nx = x + dx[next_dir]
        ny = y + dy[next_dir]
        if nx in N and ny in N and smell_graph[nx][ny]==0:
          next_shark_info.append(((nx,ny),next_dir))
          move = True
          break

      if not move:
        for next_dir in dir_priority[shark_num][curr_dir]:
          nx = x + dx[next_dir]
          ny = y + dy[next_dir]
          if nx in N and ny in N and smell_graph[nx][ny][0]==shark_num:
            next_shark_info.append(((nx,ny),next_dir))
            break

  shark_info = next_shark_info

def shark_remove():
  global shark_info
  temp = shark_info[:]
  for shark_num,(shark_loc,_) in enumerate(shark_info):
    for num,(loc,_) in enumerate(shark_info[shark_num+1:]):
      if shark_loc==loc:
        temp[shark_num+1+num] = (-1,-1)

  shark_info = temp

def solution():
  global shark_info
  global smell_graph
  global dir_priority
  time = 0

  while True:
    # 시간 증가
    time += 1

    # 냄새 그래프 시간갱신
    for i in range(n):
      for j in range(n):
        t = smell_graph[i][j]
        if type(t)==tuple:
          if t[1]-1>0:
            smell_graph[i][j] = (t[0],t[1]-1)
          elif t[1]==1:
            smell_graph[i][j] = 0

    # 현재 위치에 상어 냄새 남김
    for shark_num,(shark_loc,_) in enumerate(shark_info):
      if shark_loc!=-1:
        x,y = shark_loc
        smell_graph[x][y] = (shark_num,k)

    # 다음 위치로 상어 이동
    shark_move()

    # 겹치면 상어 없애기
    shark_remove()

    # 0번상어 밖에 안남으면 종료
    if shark_info[1:]==[(-1,-1)] * (m-1):
      break

    # 1000초 지났는데 안끝났으면 종료
    if time==1000:
      time = -1
      break

  return time

n,m,k = map(int,input().split())

dx = [-1,1,0,0]
dy = [0,0,-1,1]
N = list(range(n))

# 몇 번 상어의 냄새인지, 몇 초 남았는지 나타내는 그래프
smell_graph = []

# 각 상어의 현재 위치,방향 저장하는 리스트
shark_info = [-1] * m

for _ in range(n):
  smell_graph.append(list(map(int,input().split())))

shark_dir = list(map(int,input().split()))
for i in range(n):
  for j in range(n):
    a = smell_graph[i][j]
    if a>0:
      smell_graph[i][j] = (a-1,k)
      shark_info[a-1] = ((i,j),shark_dir[a-1]-1)

# n번 상어가 현재방향 m일 때 방향우선순위 : dir_priority[n][m]
dir_priority = [[] for _ in range(m)]

for i in range(m):
  ishark_dir = []
  for _ in range(n):
    a = list(map(int,input().split()))
    ishark_dir.append([p-1 for p in a])
  dir_priority[i] = ishark_dir

print(solution())
```
