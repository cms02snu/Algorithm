```python
# 13460

n,m = 5,5
graph = [['#', '#', '#', '#', '#'],
  ['#', '.', '.', 'B', '#'],
  ['#', '.', '#', '.', '#'],
  ['#', 'R', 'O', '.', '#'],
  ['#', '#', '#', '#', '#']]

dx = [-1,1,0,0]
dy = [0,0,-1,1]

def move_rb(graph,r_loc,b_loc,dir):
  x,y = r_loc
  while True:
    nx = x + dx[dir]
    ny = y + dy[dir]
    if graph[nx][ny]=='#':
      final_r_loc = (x,y)
      break
    elif graph[nx][ny]=='O':
      final_r_loc = (-1,-1)
      break
    else:
      x,y = nx,ny

  x,y = b_loc
  while True:
    nx = x + dx[dir]
    ny = y + dy[dir]
    if graph[nx][ny]=='#':
      final_b_loc = (x,y)
      break
    elif graph[nx][ny]=='O':
      final_b_loc = (-1,-1)
      break
    else:
      x,y = nx,ny

  if final_b_loc==(-1,-1):
    return -1,-1,-1
  elif final_r_loc==(-1,-1):
    return 1,-1,-1
  elif final_b_loc==final_r_loc:
    if dir==0:
      if b_loc[0]<r_loc[0]:
        final_r_loc = (final_r_loc[0]+1,final_r_loc[1])
      else:
        final_b_loc = (final_b_loc[0]+1,final_b_loc[1])
    elif dir==1:
      if b_loc[0]<r_loc[0]:
        final_b_loc = (final_b_loc[0]-1,final_b_loc[1])
      else:
        final_r_loc = (final_r_loc[0]-1,final_r_loc[1])
    elif dir==2:
      if b_loc[1]>r_loc[1]:
        final_b_loc = (final_b_loc[0],final_b_loc[1]+1)
      else:
        final_r_loc = (final_r_loc[0],final_r_loc[1]+1)
    elif dir==3:
      if b_loc[1]<r_loc[1]:
        final_b_loc = (final_b_loc[0],final_b_loc[1]-1)
      else:
        final_r_loc = (final_r_loc[0],final_r_loc[1]-1)

  return 0,final_r_loc,final_b_loc

def move_r_b(graph,r_loc,b_loc,dir):
  x,y = r_loc
  while True:
    nx = x + dx[dir]
    ny = y + dy[dir]
    if graph[nx][ny]=='#':
      next_r_loc = (x,y)
      break
    elif graph[nx][ny]=='O':
      return 1,-1,-1
    else:
      x,y = nx,ny

  x,y = b_loc
  while True:
    nx = x + dx[dir]
    ny = y + dy[dir]
    if graph[nx][ny]=='#':
      next_b_loc = (x,y)
      break
    elif graph[nx][ny]=='O':
      return -1,-1,-1
    else:
      x,y = nx,ny

  return 0,next_r_loc,next_b_loc

def incline(graph,dir):
  for i in range(n):
    for j in range(m):
      if graph[i][j]=='R':
        r_loc = (i,j)
      elif graph[i][j]=='B':
        b_loc = (i,j)

  # 위나 아래로 기울이는 경우
  if dir<=1:
    # R과 B가 같은 열에 있는 경우
    if r_loc[1]==b_loc[1]:
      check,next_r_loc,next_b_loc = move_rb(graph,r_loc,b_loc,dir)
    # R과 B가 다른 열에 있는 경우
    else:
      check,next_r_loc,next_b_loc = move_r_b(graph,r_loc,b_loc,dir)
  # 좌나 우로 기울이는 경우
  else:
    # R가 B가 같은 행에 있는 경우
    if r_loc[0]==b_loc[0]:
      check,next_r_loc,next_b_loc = move_rb(graph,r_loc,b_loc,dir)
    # R과 B가 다른 행에 있는 경우
    else:
      check,next_r_loc,next_b_loc = move_r_b(graph,r_loc,b_loc,dir)

  return check,next_r_loc,next_b_loc

def dfs(graph,count):
  '''
만약 check이 0이면 다음 dfs 할 때 현재위치와 다음위치 바꾸고 다시 복구하는 식으로
진행해야됨
incline 함수에서 다음위치를 결과값으로 받아야 됨
  '''

  global result
  if count==10:
    return

  for i in range(n):
    for j in range(m):
      if graph[i][j]=='R':
        curr_r_loc = (i,j)
      elif graph[i][j]=='B':
        curr_b_loc = (i,j)

  for dir in range(4):
    check,next_r_loc,next_b_loc = incline(graph,dir)
    xr,yr = curr_r_loc
    xb,yb = curr_b_loc
    if check==0:
      nxr,nyr = next_r_loc
      nxb,nyb = next_b_loc
      graph[xr][yr] = '.'
      graph[xb][yb] = '.'
      graph[nxr][nyr] = 'R'
      graph[nxb][nyb] = 'B'
      dfs(graph,count+1)
      graph[nxr][nyr] = '.'
      graph[nxb][nyb] = '.'
      graph[xr][yr] = 'R'
      graph[xb][yb] = 'B'

    if check==1:
      result = min(result,count+1)

def solution(n,m,graph):
  global result
  result = 11

  dfs(graph,0)

  if result==11:
    return -1

  return result

n,m = map(int,input().split())
graph = []
for _ in range(n):
  graph.append(list(input()))

print(solution(n,m,graph))
```
